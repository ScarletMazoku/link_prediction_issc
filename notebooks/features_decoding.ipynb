```python
#Сохраняем в файл
with open('result.json', 'w', encoding='utf-8') as result_file:
    json.dump(founded, result_file, ensure_ascii=False, indent=4)

#Удаляем из data и из publications найденные статьи
for key in founded.keys():
  data.pop(key)

for el in founded.values():
  publications = publications[publications != el]
```

```python
from nltk.metrics import edit_distance
import re

# Функция для очистки строки
def clean_text(text):
    # Удаляем e-mail, адреса и шум
    text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b', '', text)  # Удаляем e-mail
    text = re.sub(r'[^\w\s]', '', text)  # Удаляем пунктуацию
    text = re.sub(r'\s+', ' ', text)  # Убираем лишние пробелы
    text = str.lower(text)
    return text.strip()

# Функция для выделения названия
def extract_core_text(file_content):
    # Берем только первые 100-150 символов (чаще название статьи в начале)
    core_text = ' '.join(file_content.split()[:20])  # Учитываем первые слова
    return clean_text(core_text)

publications1=publications
data1 = data

for el in publications1:
  el = extract_core_text(el)

for key, el2 in data1.items():
  el2 = extract_core_text(el2)

founded = {}
for key, el2 in data1.items():
  best_match = None
  best_distance = float('inf')
  for el in publications1:
    distance = edit_distance(el2, el)
    if distance < best_distance:
      best_match = el
      best_distance = distance
  print(f'{key}: {best_match}')
  founded[key] = el
  #удаляем из publications1 el
  publications1 = publications1[publications1 != best_match]
print(len(founded))
```

```python
import os

files_dict = {}

directory_path = '/content/drive/MyDrive/NIR_kurs2/ICCS_files/ICCS_files'

for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory_path, filename)
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                first_200_chars = file.read(200)
                files_dict[filename] = first_200_chars
        except Exception as e:
            print(f"Ошибка при обработке файла {filename}: {e}")

files_dict['ICCS_2021_298.txt']
```
